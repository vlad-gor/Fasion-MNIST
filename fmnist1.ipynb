```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
from sklearn.linear_model import LogisticRegression

%matplotlib inline
```

```python
# В этом блокноте - обучение логистической регрессии с использованием модуля sklearn
# Во второй части - обучение логистической регрессии и нейронных сетей с использованием tensorflow

train = pd.read_csv('fashion-mnist_train.csv')
test = pd.read_csv('fashion-mnist_test.csv')

x_train = train.loc[:, train.columns != 'label'] 
y_train = train['label']

x_test = test.loc[:, test.columns != 'label']
y_test = test['label']

x_train = x_train.astype('float32')
x_test = x_test.astype('float32')
x_train, x_test = x_train / 255.0, x_test / 255.0
```

```python
enc = OneHotEncoder()

y_train = enc.fit_transform(np.reshape(np.array(y_train), (-1,1))).toarray()
y_test = enc.transform(np.reshape(np.array(y_test), (-1,1))).toarray()
```

```python
print(x_test.shape)
print(x_train.shape)
print(y_test.shape)
print(y_train.shape)
```

```python
# 1. Обучение логистической регрессии

LogReg = LogisticRegression()

lg_X_train = np.array(x_train)
lg_X_test = np.array(x_test)
y_train = train['label']
y_test = test['label']
```

```python
f = LogReg.fit(lg_X_train, y_train)
f.get_params()
```

```python
predictions = LogReg.predict(lg_X_test)
print(predictions)
```

```python
score = LogReg.score(lg_X_test, test['label'])
print('\nТочность на проверочных данных, используя Logistic Regression: {}'.format(score))
```
